## Calculus

### Use Python like a calculuator

Approximate $f'(1)$ of $f(x) = x^2$.

```python
x = 1
h = 0.01
df = ((x + h)**2 - x**2)/h
print(df)
```

Approximate $f'(1)$ of $f(x) = x^x$.

```python
x = 1
h = 0.00001
df = ((x + h)**x - x**x)/h
print(df)
```

Approximate $\lim_{x \to 1} \frac{x^2 - 1}{x - 1}$.

```python
x = 1.0001
L = (x**2 - 1)/(x - 1)
print(L)
```

Approximate zero of $f(x) = x^5 + x - 1$ near 1/2 using Newton's method.

```python
x0 = 1/2
x1 = x0 - (x0**5 + x0 - 1)/(5*x0**4 + 1)
print(x1)
```

```python
x2 = x1 - (x1**5 + x1 - 1)/(5*x1**4 + 1)
print(x2)
```

```python
x3 = x2 - (x2**5 + x2 - 1)/(5*x2**4 + 1)
print(x3)
```

### Define Functions and Use NumPy Functions

```python
f = lambda x: x**5 + x - 1
df = lambda x: 5*x**4 + 1
x0 = 1/2
x1 = x0 - f(x0)/df(x0)
print(x1)
```

Approximate the limit $\lim_{x \to 0} \frac{\sin(x)}{x}$.

```python
x = 0.001
L = sin(x)/x
print(L)
```

### Graphical Methods

Use the `graph` function to plot the graph of $f(x) = x e^{-x^2}$ over $[-5,5]$.

```python
f = lambda x: x*exp(-x**2)
graph(f,-5,5)
```

```python
f1 = lambda x: x*exp(-x**2)
f2 = lambda x: x*exp(-x**4)
graph(f1,-2,2)
graph(f2,-2,2)
```

Plot the tangent line of $f(x) = \frac{x^{3}+x-1}{2x^{2}+1}$ at $x=0$.

```python
f = lambda x: (x**3 + x - 1)/(2*x**2 + 1)
x = 0
h = 0.0001
m = (f(x + h) - f(x))/h
b = f(x)
```

```python
L = lambda x: m*x + b
```

```python
graph(f,-1,1)
graph(L,-1,1)
point(0,b)
```
